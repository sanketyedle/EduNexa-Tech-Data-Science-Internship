# Scatter Plot for Sepal Length and Sepal Width

```python
df.plot(kind='scatter',
		x='sepal_length',
		y='sepal_width',
		color='red')
```

# Scatter Plot 'Bar'

```python
df.plot(kind='bar',
		x='sepal_length',
		y='sepal_width',
		color='blue')
```

# Pie Plot to Get the Frequency of the Three Species of the IRIS Dataset

```python
iris = pd.read_csv("IRIS.csv")
ax=plt.subplots(1,1,figsize=(10,8))
iris['species'].value_counts().plot.pie(explode=[0.1,0.1,0.1],autopct='%1.1f%%',shadow=True,figsize=(10,8))
plt.title("Iris Species %")
plt.show()
```

# Box Plot for IRIS Dataset

```python
plt.figure(figsize = (10, 7)) 
df.boxplot() 
```

# Histogram for Sepal Width

```python
plt.figure(figsize = (10,10)) 
x = df["sepal_width"] 
  
plt.hist(x, bins = 20, color = "green") 
plt.title("Sepal Widthth in cm") 
plt.xlabel("sepal_wength") 
plt.ylabel("Count") 
```

# Histogram for Petal Width

```python
plt.figure(figsize = (10, 7)) 
x = df["petal_width"] 
  
plt.hist(x, bins = 20, color = "blue") 
plt.title("Petal Width in cm") 
plt.xlabel("petal_width") 
plt.ylabel("Count") 
```

# Histogram for Petal Lenght

```python
plt.figure(figsize = (10, 7)) 
x = df["petal_length"] 
  
plt.hist(x, bins = 20, color = "red") 
plt.title("Petal Length in cm") 
plt.xlabel("petal_length") 
plt.ylabel("Count") 
```

# Scatter Plot for Petal Length and Petal Width

```python
df.plot(kind='scatter',
		x='petal_length',
		y='petal_width',
		color='orange')
```

# Line Chart Plotting

```python
plt.figure(figsize=(14,6))
plt.title("IRIS Dataset Infromation")
sns.lineplot(data=df)
```

# Seaborn Jointplot

```python
sns.jointplot(x="sepal_length", y="sepal_width", data=df, size=5)
```

# Violin Plot

```python
sns.violinplot(x="species", y="sepal_width", palette="husl", data=df)
```

# Swarm Plot

```python
import warnings
warnings.filterwarnings('ignore')
sns.swarmplot(x="species", y="petal_length", data=df, color="blue", alpha=.9);
```
