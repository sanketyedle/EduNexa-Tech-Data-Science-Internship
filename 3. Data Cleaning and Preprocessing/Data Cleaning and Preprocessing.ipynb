# Filling Missing/Null Values By Using Various Techniques
## Fill Missing or Null Values with 0 (Temperory)

```python
df.fillna(0)
```

## df.fillna(0) Fills Null Values Temporarily, If we display the data set it shows original data set with missing values (Output Given Below)

```python
df
```

## Fill Missing Values By Mean (Each Column)

```python
df["Math_score"] = df["Math_score"].fillna(df["Math_score"].mean())
```

```python
df["Writing_score"] = df["Writing_score"].fillna(df["Writing_score"].mean())
```

## To Check Null Values in Data Set (Column 'Math_score' & 'Writing_score')

```python
df.isnull().sum()
```

## Fill Missing Values By Median

```python
df["Reading_score"] = df["Reading_score"].fillna(df["Reading_score"].median())
```

```python
df.isnull().sum()
```

# Outliers
## Find Outliers By Boxplot

```python
col=['Math_score','Reading_score','Writing_score','Placement_score']
```

```python
df.boxplot(col)
```

## Z Score on Math_score Outlier Detection

```python
z=stats.zscore(df['Math_score'])
z
```

## Find Exact Location of Outlier

```python
threshold=3
sample_outliers=np.where(z>threshold)
sample_outliers
```

# Quantiles
## Find outliers using Quantile 25%

```python
Q1=df.Math_score.quantile(0.25)
Q1
```

## Find outliers using Quantile 25%

```python
Q3=df.Math_score.quantile(0.75)
Q3
```

## Find Inter Quartile Range

```python
IQR = Q3-Q1
IQR
```

## Find Lower Limit

```python
lower_limit=Q1-1.5*IQR
lower_limit
```

## Find Upper Limit

```python
upper_limit=Q3+1.5*IQR
upper_limit
```

## Display Outlier For Specific Column

```python
sample_outliers=df['Math_score'][(df['Math_score']>upper_limit) | (df['Math_score']<lower_limit)]
```

```python
sample_outliers
```

## Find Mean For Specific Column

```python
mean1 = df['Math_score'].mean()
mean1
```

## Remove/Replace Outliers by Mean

```python
df['Math_score']=np.where((df['Math_score']>upper_limit) | (df['Math_score']<lower_limit), mean1, df['Math_score'])
```

```python
df['Math_score']
```

# Scaling :Min-Max Algoorithm
## Apply Data Transformations Using Min Max

```python
df_min_max_scaled = df.copy()
for column in df_min_max_scaled.columns:
    df_min_max_scaled[column]=(df_min_max_scaled[column] - df_min_max_scaled[column].min()) 
    / (df_min_max_scaled[column].max()-df_min_max_scaled[column].min())
```

```python
df_min_max_scaled
```

## Distribution Plot (Graph)

```python
import warnings
warnings.filterwarnings('ignore')
sns.set_style('whitegrid')
sns.displot(df_min_max_scaled["Math_score"], kde=True, bins=15 ,linewidth = 0.5)
sns.displot(df_min_max_scaled["Reading_score"], kde=True, bins=15,linewidth = 0.5)
sns.displot(df_min_max_scaled["Writing_score"], kde=True, bins=15,linewidth = 0.5)
sns.displot(df_min_max_scaled["Placement_score"], kde=True, bins=15,linewidth = 0.5)
sns.displot(df_min_max_scaled["Club_Join_Date"], kde=True, bins=15,linewidth = 0.5)
sns.displot(df_min_max_scaled["Placement_Offer_Count"], kde=True, bins=15 ,linewidth = 0.5)

plt.tight_layout() 
plt.show()
```
